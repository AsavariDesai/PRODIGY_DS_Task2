```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('.\Documents\Internship\insurance.csv') 
df.head()  	
```

```python
df.head()
df.info()
df.isnull().sum()
```

```python
df['age'].fillna(df['age'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Gender'] = df['Gender'].map({'male': 0, 'female': 1})
df['smoker'] = df['smoker'].map({'yes': 0, 'no': 1})
df.head()
```

```python
# Count the number of passengers in each category
Gender_counts = df['Gender'].value_counts()
smoker_counts = df['smoker'].value_counts()
print(Gender_counts)
print(smoker_counts)

# Plot histograms of the 'Age' and 'Fare' columns
sns.histplot(data=df, x="age", kde=False, bins=20)
plt.title("Distribution of Ages")
plt.show()

sns.histplot(data=df, x="smoker", kde=False, bins=20)
plt.title("Distribution of Smoker")
plt.show()
```

```python
plt.figure(figsize=(10, 6))

male_df = df[df['Gender'] == 0]
female_df = df[df['Gender'] == 1]
print(male_df)
print(female_df)

plt.bar(["Male", "Female"], [male_df['age'].mean(), female_df['age'].mean()])
plt.xlabel("Gender")
plt.ylabel("Age")
plt.title("Average Age of People by Gender")
plt.show()
```

```python
sns.scatterplot( x="charges", y='region', data=df,
                hue='Gender', size='bmi')
 
# Placing Legend outside the Figure
plt.legend(bbox_to_anchor=(1, 1), loc=2)
 
plt.show()
```
